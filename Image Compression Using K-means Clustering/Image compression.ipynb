```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.utils import shuffle
from PIL import Image

# Step 1: Load and preprocess the image
def load_image(image_path):
    image = Image.open(image_path)
    image = np.array(image)
    return image

# Step 2: Compress the image with K-means
def compress_image_kmeans(image, n_colors=16):
    # Reshape the image into (num_pixels, 3)
    width, height, _ = image.shape
    image_array = image.reshape(-1, 3)

    # Sample pixels to speed up the clustering process
    image_sample = shuffle(image_array, random_state=0)[:1000]
    
    # Apply K-means clustering
    kmeans = KMeans(n_clusters=n_colors, random_state=0)
    kmeans.fit(image_sample)

    # Replace each pixel with the color of its cluster's centroid
    labels = kmeans.predict(image_array)
    compressed_image_array = kmeans.cluster_centers_[labels].astype(int)
    
    # Reshape back to the original image dimensions
    compressed_image = compressed_image_array.reshape(width, height, 3)
    return compressed_image

# Step 3: Plot the original and compressed images
def plot_comparison(original, compressed):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    ax1.imshow(original)
    ax1.set_title("Original Image")
    ax1.axis("off")

    ax2.imshow(compressed)
    ax2.set_title("Compressed Image")
    ax2.axis("off")
    plt.show()

# Load and process the image
image_path = "D:\Downloads feom c\Downloads\data science logo.jpg"  # Replace with your image path
image = load_image(image_path)

# Compress the image using K-means
compressed_image = compress_image_kmeans(image, n_colors=16)

# Display the result
plot_comparison(image, compressed_image)
```
